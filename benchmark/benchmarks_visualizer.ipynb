```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/all_benchmark_data.csv')
```

```python
speed_data = data[data['metric_name'] == 'speed']
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.lineplot(data=speed_data, x='x_value', y='y_value', hue='kernel_provider', marker='o', palette='tab10')
plt.ylabel('Speed (ms)')
plt.legend(title='Kernel Provider')
plt.xlabel('Vocab Size')
plt.tight_layout()
```
